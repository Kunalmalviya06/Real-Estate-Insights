```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Downloads/flats.csv")
```

```python
flats_df = data
```

```python
flats_df.drop(columns=['link'], inplace=True)
```

```python
# Clean price Column

def clean_price(price):
    if isinstance(price, str):
        price = price.strip()
        if 'Lac' in price:
            return float(price.replace(' Lac', '').replace(',', '')) * 100000
        elif 'Crore' in price:
            return float(price.replace(' Crore', '').replace(',', '')) * 10000000
    return 0 

flats_df['price'] = flats_df['price'].apply(clean_price)
print(flats_df[['price']].head())
print(flats_df.info())
```

```python
# Q1 How many flats are listed in each society?

society_counts = flats_df['society'].value_counts().head(10)
print(society_counts)
```

```python
# Visualization

plt.figure(figsize=(14, 8))
sns.barplot(x=society_counts.values, y=society_counts.index, palette='coolwarm')

plt.title('Top 10 Societies with Most Flats Listed', fontsize=20, weight='bold')
plt.xlabel('')
plt.ylabel('')

for index, value in enumerate(society_counts):
    plt.text(value, index, str(value), fontsize=12, color='black', va='center')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Q2 What is the average price of 2 BHK flats?

average_price_2_bhk = flats_df[flats_df['bedRoom'] == '2 Bedrooms']['price'].mean()
print("The average price of 2 BHK flats is: ₹{:.2f}".format(average_price_2_bhk))
```

```python
# Visualization

average_price_by_bedroom = flats_df.groupby('bedRoom')['price'].mean()
plt.figure(figsize=(10, 6))
bars = flats_df.groupby('bedRoom')['price'].mean().plot(kind='bar', color='steelblue')

plt.title('Average Price of Flats by Number of Bedrooms', fontsize=16)
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0)
plt.grid(axis='y')

for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():,.0f}', 
             ha='center', va='bottom')

plt.show()
```

```python
# Q5 Determine the average area of flats for each type of facing (e.g., North, South).

average_area_by_facing = flats_df.groupby('facing')['area'].mean()
print(average_area_by_facing)
```

```python
# Clean Price Column

def cleaned_price(price):
    if isinstance(price, str):
        price = price.strip()
        if 'Lac' in price:
            return float(price.replace(' Lac', '').replace(',', '')) * 100000
        elif 'Crore' in price:
            return float(price.replace(' Crore', '').replace(',', '')) * 10000000
    return 0

houses_df['price'] = houses_df['price'].apply(cleaned_price)
houses_df.info()
```

```python
# Q1 What is the most common number of floors in the listed houses?

most_common_floors = houses_df['noOfFloor'].mode()[0]
print(most_common_floors)
```

```python
# Q2 Calculate the average price of houses with more than 3 bedrooms.

average_price = houses_df[houses_df['bedRoom'] > 3]['price'].mean()
print(average_price)

average_price_by_bedroom = houses_df[houses_df['bedRoom'] > 3].groupby('bedRoom')['price'].mean().head()
print(average_price_by_bedroom)

```

```python
# Vuisualization 

plt.figure(figsize=(10, 6))
average_price_by_bedroom.plot(kind='bar', color='steelblue')

plt.title('Average Price of Houses by Bedroom Count', fontsize=16)
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0)
plt.grid(axis='y')

for index, value in enumerate(average_price_by_bedroom):
    plt.text(index, value, f'₹ {value:,.0f}', ha='center', va='bottom')

plt.show()
```

```python
# Q3 Compare the average areas of furnished vs. unfurnished houses.

houses_df[['furnishDetails']]
```

```python
# Q3 Compare the average areas of furnished vs. unfurnished houses.


def determine_furnishing(furnish_details):
    if pd.isna(furnish_details):
        return 'Unfurnished'
    
    items = eval(furnish_details)  # Convert string representation to list
    furnished_items = ['Wardrobe', 'Fan', 'Fridge', 'Bed', 'Geyser', 'Light', 'AC']

    # Check if any furnished item is present in the furnishDetails
    for item in items:
        if any(furnished_item in item for furnished_item in furnished_items):
            return 'Furnished'
    
    return 'Unfurnished'

houses_df['status'] = houses_df['furnishDetails'].apply(determine_furnishing)
print(houses_df[['furnishDetails', 'status']].head())
```

```python
# Q3 Compare the average areas of furnished vs. unfurnished houses.

# Clean Area Column 
houses_df['area_value'] = houses_df['areaWithType'].str.extract('(\d+\.?\d*)')
houses_df['area_type'] = houses_df['areaWithType'].str.extract('([A-Za-z ]+)')
houses_df['area_value'] = houses_df['area_value'].astype(float)
print(houses_df[['areaWithType', 'area_type', 'area_value']].head())

```

```python
# Q3 Compare the average areas of furnished vs. unfurnished houses.

average_area_comparison = houses_df.groupby('status')['area_value'].mean()
print(average_area_comparison)
```

```python
# Visualization

plt.figure(figsize=(8, 6))
plt.pie(average_area_comparison, labels=average_area_comparison.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Blues", len(average_area_comparison)))

centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Average Area Comparison: Furnished vs Unfurnished Houses', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
# Key Performance Indicator (KPI)

average_price_flats = flats_df['price'].mean()
average_price_houses = houses_df['price'].mean()
average_area_flats = flats_df['area'].mean()
average_area_houses = houses_df['area_value'].mean()

print("Flats Data KPIs:")
print(f"Average Price per Flat: ₹{average_price_flats:,.2f}")
print(f"Average Area per Flat: {average_area_flats:.2f} sq.ft")
```

```python
print("\nHouses Data KPIs:")
print(f"Average Price per House: ₹{average_price_houses:,.2f}")
print(f"Average Area per House: {average_area_houses:.2f} sq.ft")
```
